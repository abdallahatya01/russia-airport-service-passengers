```python
dm = df.iloc[:,1:14]
months = dm.columns[1:].tolist()
months
```

```python
dm = dm.melt(id_vars =  'Year', value_vars = months)
dm.columns = ['Year', 'Month', 'Passengers']
dm = dm.groupby(['Year', 'Month']).sum().reset_index()
dm
```

```python
plt.figure(figsize = (20,10))

sns.lineplot(data = dm, x = 'Year', y = 'Passengers', color = 'r')
plt.ylabel('Passengers in million', fontsize=  16)
plt.title("passenger's active by years", fontsize = 19)
```

```python
plt.figure(figsize = (20,10))
sns.barplot(data = dm, x = 'Month', y = dm[dm.Year==2009].Passengers, color = 'r')
plt.title("passenger's active in 2009", fontsize = 19)
```
